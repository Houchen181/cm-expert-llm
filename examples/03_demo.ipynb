```python
from cmp_expert.serve.retriever import HybridRetriever

# Create retriever
retriever = HybridRetriever()

# Add physics documents
docs = [
    {"id": "sc1", "text": "Superconductivity has zero resistance below Tc"},
    {"id": "ti1", "text": "Topological insulators conduct on surface only"}
]
retriever.add_documents(docs)
print(f"Indexed {len(docs)} documents")
```

```python
# Search
results = retriever.search("superconductivity", k=1)
print(f"Found: {results[0]['text'] if results else 'Nothing'}")
```
