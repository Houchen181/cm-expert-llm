# CM-Expert-LLM: Training with LoRA

This notebook demonstrates how to fine-tune a base LLM for condensed matter physics using LoRA (Low-Rank Adaptation).

## Overview

The training pipeline:
1. Loads preprocessed data from `data/processed/sft.jsonl`
2. Configures LoRA parameters (rank, alpha, target modules)
3. Optionally performs Domain-Adaptive Pretraining (DAPT)
4. Trains LoRA adapters on instruction-tuning data
5. Saves adapter weights for deployment

## Step 3: Training Configuration Details

### LoRA Parameters

- **Rank (r)**: Dimension of low-rank matrices (typically 8-64)
- **Alpha**: Scaling factor for LoRA updates (typically 2x rank)
- **Target Modules**: Which transformer layers to adapt (q_proj, v_proj, etc.)

### Domain-Adaptive Pretraining (DAPT)

DAPT is an optional first stage that continues pretraining on domain corpus before instruction tuning. This helps the model learn domain-specific vocabulary and concepts.

```python
# Print detailed LoRA config
lora_config = config.get('lora', {})
print("LoRA Configuration:")
print(f"  Rank (r): {lora_config.get('r')}")
print(f"  Alpha: {lora_config.get('alpha')}")
print(f"  Dropout: {lora_config.get('dropout')}")
print(f"  Target Modules: {lora_config.get('target_modules')}")

print("\nTraining Hyperparameters:")
training = config.get('training', {})
print(f"  Epochs: {training.get('epochs')}")
print(f"  Learning Rate: {training.get('lr')}")
print(f"  Batch Size: {training.get('batch_size')}")
print(f"  Max Length: {training.get('max_length')}")
```
